## MY CLEANING FUNCTION

```python
import numpy as np
import pandas as pd

def clean_df(df):
    ca## CLEANING FUNCTIONtegorical = df.select_dtypes(include=object)
    numerical = df.select_dtypes(include=np.number)

    # Cleaning Weight column
    df['Weight'] = df['Weight'].str.replace('lbs', '')  # delete the units lbs
    # convert it to a number
    df['Weight'] = pd.to_numeric(df['Weight'])
    df = df.rename(columns={'Weight': 'Weight(kg)'})  # add lbs in the title
    df['Weight(kg)'] = df['Weight(kg)'] * 0.45359237
    df.sort_values(by="OVA", ascending=False, inplace=True)
    df['Contract'] = df['Contract'].str.replace("~", " ")  # replace the symbol using space
    df[['Contract_starts', 'Contract_ends']] = df['Contract'].str.extract(r'(\d{4})\s*(\d{4})')  # extract by 2 new columns
    df[['Value', 'Wage', 'Release Clause']] = df[['Value', 'Wage', 'Release Clause']].replace("€", regex=True)

    def convert_value(value):
        if isinstance(value, str):  # Verificar si el valor es una cadena
            if 'M' in value:
                return float(value.replace('M', '').replace('€', '')) * 1000000
            elif 'K' in value:
                return float(value.replace('K', '').replace('€', '')) * 1000
            elif 'B' in value:
                return float(value.replace('B', '').replace('€', '')) * 1000000000
            elif '€' in value:
                return float(value.replace('€', ''))
            else:
                return float(value)
        else:
            return value

    df['Value'] = df['Value'].replace("€", "", regex=True).apply(convert_value)
    df['Wage'] = df['Wage'].replace("€", "", regex=True).apply(convert_value)
    df['Release Clause'] = df['Release Clause'].replace("€", "", regex=True).apply(convert_value)

    df['W/F'] = df['W/F'].str.replace('★', '')
    df['SM'] = df['SM'].str.replace('★', '')
    df['IR'] = df['IR'].str.replace('★', '')

    for index in range(73, 100):
        column_name = df.columns[index]
        column_data = df[column_name]

        for i, value in enumerate(column_data):
            # Split values based on '+' or '-'
            value_str = str(value)
            split_vals = value_str.split('+') if '+' in value_str else value_str.split('-')

            # Perform addition or subtraction
            if len(split_vals) > 1:
                operation_result = int(split_vals[0]) + int(split_vals[1]) if '+' in str(value) else int(split_vals[0]) - int(split_vals[1])
            else:
                operation_result = int(split_vals[0])

        df.at[i, column_name] = operation_result
        null_values = df.isnull().sum()

    # numerical columns
    numeric_columns = df.select_dtypes(include='number').columns
    df[numeric_columns] = df[numeric_columns].fillna(0).astype(int)

    for col in numeric_columns:
        df[col] = df[col].fillna(df[col].median(numeric_only=False)).astype(int)

    # categorical columns
    categoric_columns = df.select_dtypes(exclude='number').columns
    for col in categoric_columns:
        df[col] = df[col].fillna('F').astype('category')

    return df
```
